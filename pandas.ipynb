```python
n=int(input())
li = list(map(int, input(). split()))

mean=sum(li) //n



max_ =max(li)
min_ =min(li)

maxID = None
minID = None

for i in range(n):
  if li[i] ==max_:
    maxID = i
  if li[i] == min_:
    minID = i

trans =max_ - mean

li[maxID] = li[maxID] - trans
li[minID] = li[minID] +trans

for v in li:
  print(v, end=" ")
```

```python
"""
終極密碼遊戲
電腦隨機產生一個密碼
由使用者去猜
猜到密碼就結束遊戲
"""

import random
ans = random.randint(1, 100)
left = 1
right = 100
while True:
  guess = int(input("請隨機猜數字: "))
  if guess == ans:
    print("恭喜猜中")
    break
  elif guess > ans:
    right = guess
  elif guess < ans:
    left = guess
  print(f"{left} - {right}")
```

```python
n = int(input())
left, right, uturn = 0, 0, 0
cx, cy = 0, 0
dir = 0 # 0: 右, 1: 下, 2: 左, 3: 上
for i in range(n):
  nx, ny = map(int, input().split())
  if nx > cx:
    if dir == 1:
      left += 1
    elif dir == 2:
      uturn += 1
    elif dir == 3:
      right += 1
    dir = 0
  if ny < cy:
    if dir == 0:
      right += 1
    elif dir == 2:
      left += 1
    elif dir == 3:
      uturn += 1
    dir = 1
  if nx < cx:
    if dir == 0:
      uturn += 1
    elif dir == 1:
      right += 1
    elif dir == 3:
      left += 1
    dir = 2
  if ny > cy:
    if dir == 0:
      left += 1
    elif dir == 1:
      uturn += 1
    elif dir == 2:
      right += 1
    dir = 3
  cx, cy = nx, ny

print(left, right, uturn)

```

```python
n, D = map(int, input().split())
li = list(map(int, input().split()))
ans = 0
buy = li[0]
sell = 0
for i in range(1, n):
  price = li[i]
  if buy > 0:
    if price >= buy + D:
      ans += price - buy
      sell = price
      buy = 0
  else:
    if price <= sell - D:
      buy = price
      sell = 0
print(ans)

```

```python
class Warrior:

  # Constructor 建構子
  def __init__(self, name, gender, weapon = "bat"): #想像你設計了一個「遊戲角色」的製造藍圖（類別 Class）。每當玩家創建一個新角色（物件 Object）時，這個新角色總該有名字、血量和職業吧？

                               # __init__ 就是那個「新角色一出生，就自動幫他設定好名字和血量」的啟動按鈕。
    # 共同屬性
    self.level = 1
    self.exp = 0
    # 客製化屬性
    self.n = name
    self.g = gender
    self.w = weapon
    print(f"建立了戰士: {self.n}")

  # 方法
  def attack(self, enemy):
    print(f"{self.n} 用 {self.w} 攻擊 {enemy.n}")

  def levelUP(self):
    self.level += 1
    self.exp = 0
    print(f"{self.n} 升級，現在是 {self.level} 級")
```

```python
# 物件實體化
w1 = Warrior(name="Lucas", gender = "Male")
w2 = Warrior(name="Allie", gender = "Female", weapon="Sword")

print(f"w1 戰士姓名: {w1.n}, w1 戰士武器: {w1.w}")
print(f"w2 戰士姓名: {w2.n}, w2 戰士武器: {w2.w}")

w1.attack(w2)
w2.attack(w1)
w1.levelUP()
print(w1.level, w2.level)
```
