```python
from math import radians, sin, cos, sqrt, atan2

WALK_KMH = 4.8

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    dlat, dlon = radians(lat2 - lat1), radians(lon2 - lon1)
    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    return R * 2 * atan2(sqrt(a), sqrt(1 - a))

def nearest(lat, lng, mode="借車", top=5):
    # TODO(1)：依 mode 選欄位（借車看 available_rent_bikes、還車看 available_return_bikes），
    #          只保留該欄位 > 0 的站，計算每站到 (lat,lng) 的距離與步行分鐘，
    #          由近到遠排序後回傳前 top 站，欄位需含 sna, sarea, avail, dist, walk, latitude, longitude。

    # 1. 決定目標欄位
    target_col = "available_rent_bikes" if mode == "借車" else "available_return_bikes"

    # 2. 過濾出有車子可借/有空位可還的站點
    available_df = df[df[target_col] > 0].copy()

    # 3. 計算距離（公里）與步行時間（分鐘）
    available_df["dist"] = available_df.apply(
        lambda r: haversine(lat, lng, r["latitude"], r["longitude"]), axis=1
    )
    available_df["walk"] = (available_df["dist"] / WALK_KMH) * 60

    # 4. 統一重新命名欄位名稱為 avail
    available_df["avail"] = available_df[target_col]

    # 5. 排序並取出前 top 筆
    result_df = available_df.sort_values(by="dist").head(top)

    return result_df[["sna", "sarea", "avail", "dist", "walk", "latitude", "longitude"]]
```

```python
from ipyleaflet import Map, Marker, CircleMarker, Polyline, AwesomeIcon
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import requests

# Fetch data and define df within this cell for self-containment
URL = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"
resp = requests.get(URL, timeout=10)
records = resp.json()
df = pd.DataFrame(records)

# Ensure numeric columns are correctly typed
for col in ["available_rent_bikes", "available_return_bikes", "total", "latitude", "longitude"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

m = Map(center=(25.0478, 121.5319), zoom=12, scroll_wheel_zoom=True)

# TODO(2)：把 df 裡「每個站」畫成地圖上的小圓點(CircleMarker)當底圖，
#          （迴圈跑 df.iterrows()，每站 m.add_layer(...)）
for _, row in df.iterrows():
    station_marker = CircleMarker(
        location=(row["latitude"], row["longitude"]),
        radius=3,
        color="#3498db",
        fill_color="#3498db",
        fill_opacity=0.4,
        weight=1
    )
    m.add_layer(station_marker)


mode_toggle = widgets.ToggleButtons(options=["借車", "還車"], value="借車", description="模式:")
clear_btn = widgets.Button(description="清除", button_style="warning", icon="eraser")
info_box = widgets.HTML(value="👉 滑鼠移到站點看車數；點地圖任一位置找最近 5 站")

dynamic_layers = []

def clear_dynamic(_=None):
    for layer in dynamic_layers:
        try:
            m.remove_layer(layer)
        except Exception:
            pass
    dynamic_layers.clear()
    info_box.value = "已清除"

def on_click(**kwargs):
    if kwargs.get("type") != "click":
        return
    lat, lng = kwargs["coordinates"]
    clear_dynamic()

    mode = mode_toggle.value
    # The 'nearest' function is defined in a previous cell. Assuming it's available or should be re-added if this cell is meant to be fully standalone.
    # For now, let's assume `nearest` and `haversine` are available from previous executions or imported/defined elsewhere that's not in the prompt's scope.
    result = nearest(lat, lng, mode=mode, top=5)
    kind = "可借" if mode == "借車" else "可還"

    click_marker = Marker(location=(lat, lng),
                          icon=AwesomeIcon(name="user", marker_color="red"))
    m.add_layer(click_marker); dynamic_layers.append(click_marker)

    rows = []
    # TODO(3)：把 result 的每個站標成綠色 CircleMarker，從點擊處拉一條 Polyline 到該站，
    #          並把「站名、可借/可還幾台、距離、步行分鐘」加進 rows 清單。
    #          （記得每個新圖層都要 m.add_layer(...) 並 append 到 dynamic_layers）
    for _, row in result.iterrows():
        station_loc = (row["latitude"], row["longitude"])

        # 1. 綠色 CircleMarker 標示最近站點
        green_marker = CircleMarker(
            location=station_loc,
            radius=6,
            color="#2ecc71",
            fill_color="#2ecc71",
            fill_opacity=0.9
        )
        m.add_layer(green_marker)
        dynamic_layers.append(green_marker)

        # 2. 點擊點拉線至站點
        line = Polyline(
            locations=[(lat, lng), station_loc],
            color="#2ecc71",
            weight=2,
            opacity=0.7
        )
        m.add_layer(line)
        dynamic_layers.append(line)

        # 3. 整理顯示的 HTML 字串（去除 YouBike2.0_ 的前綴字讓顯示更乾淨）
        clean_name = row["sna"].replace("YouBike2.0_", "")
        rows.append(
            f"<div style='margin-bottom: 8px; border-bottom: 1px solid #eee; padding-bottom: 5px;'>"
            f"🚶 <b>{clean_name}</b> ({row['sarea']})<br>"
            f"&nbsp;&nbsp;&nbsp;&nbsp;{kind}：<b>{int(row['avail'])}</b> 台｜距離：<b>{row['dist']:.2f}</b> km｜步行：約 <b>{row['walk']:.1f}</b> 分鐘"
            f"</div>"
        )

    info_box.value = (f"<h4>📍 最近、且{kind}的前 5 站（{mode}）</h4>"
                      + "".join(rows))


m.on_interaction(on_click)
clear_btn.on_click(clear_dynamic)

def on_mode_change(_):
    info_box.value = f"已切到「{mode_toggle.value}」，請重新點一下地圖"
mode_toggle.observe(on_mode_change, names="value")

display(widgets.HBox([mode_toggle, clear_btn]))
display(widgets.HBox([m, info_box]))
```
